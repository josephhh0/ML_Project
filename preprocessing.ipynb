```python
df_unique = df.drop_duplicates(subset=["description"])
df_unique.dropna(subset=["description"],inplace=True)
df_unique
```

```python
df_unique["description"]=df_unique["description"].astype(str)
```

```python
import re
df_unique['description'] = df_unique['description'].str.lower()      
def clean_text(x):
    x = re.sub(r'[^a-zA-Z\s]', '', x)
    return x
df_unique['description'] = df_unique['description'].apply(clean_text) 
```

```python
df_unique["description"]
```

```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

ENGLISH_STOP_WORDS = list(ENGLISH_STOP_WORDS)  #transforming ENGLISH STOP WORDS from frozenset to a list to add new stop words to it

def remove_stopwords(text):
       words = text.split()
       words = [word for word in words if word not in ENGLISH_STOP_WORDS]
       return ' '.join(words)
df_unique['description'] = df_unique['description'].apply(remove_stopwords) # Removing unecessary stop words from the reviews
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
import nltk

nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')
nltk.download("punkt_tab")
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')


lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(word):

    tag = nltk.pos_tag([word])[0][1]

    if tag.startswith('J'): return wordnet.ADJ
    elif tag.startswith('V'): return wordnet.VERB
    elif tag.startswith('N'): return wordnet.NOUN
    elif tag.startswith('R'): return wordnet.ADV
    else: return wordnet.NOUN

def lemmatize_text(text):
    words = text.split()
    words = [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in words]
    return ' '.join(words)

df_unique['description'] = df_unique['description'].apply(lemmatize_text)
df_unique['description']
```
